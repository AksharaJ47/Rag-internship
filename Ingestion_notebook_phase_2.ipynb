```python
try:
    import os

    from google.colab import userdata

    from langchain_community.document_loaders import PyPDFLoader

    from langchain_text_splitters import RecursiveCharacterTextSplitter

    from langchain_google_genai import GoogleGenerativeAIEmbeddings

    from langchain_community.vectorstores import Chroma

    print("✅ All libraries imported successfully")

except Exception as e:
    print("❌ Import Error:", e)
```

```python
try:
    from google.colab import userdata
    import os

    GOOGLE_API_KEY = userdata.get("GEMINI_KEY")

    if GOOGLE_API_KEY is None:
        raise ValueError("Gemini API Key not found in Colab Secrets")

    os.environ["GOOGLE_API_KEY"] = GOOGLE_API_KEY

    print("✅ Gemini API Key Loaded Successfully")

except Exception as e:
    print("❌ Error:", e)
```

```python
try:

    from google.colab import files

    uploaded = files.upload()

    pdf_path = list(uploaded.keys())[0]

    print("✅ Uploaded:", pdf_path)

except Exception as e:

    print("❌ Upload Error:", e)
```

```python
try:

    loader = PyPDFLoader(pdf_path)

    documents = loader.load()

    print("✅ PDF Loaded Successfully")

    print("Total Pages:", len(documents))

except Exception as e:

    print("❌ PDF Loading Error:", e)
```

```python
try:

    splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=50
    )

    chunks = splitter.split_documents(documents)

    print("✅ Chunking Completed")

    print("Total Chunks:", len(chunks))

except Exception as e:

    print("❌ Chunking Error:", e)
```

```python
try:
    !pip install -q sentence-transformers

    from langchain_community.embeddings import HuggingFaceEmbeddings

    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )

    print("✅ Hugging Face Embedding Model Loaded Successfully")

except Exception as e:
    print("❌ Embedding Error:", e)
```

```python
try:

    vectorstore = Chroma.from_documents(
        documents=chunks,
        embedding=embeddings,
        persist_directory="./chroma_db"
    )

    print("✅ Chunks Stored Successfully in ChromaDB")

except Exception as e:

    print("❌ ChromaDB Error:", e)
```

```python
print(vectorstore._collection.count())
```

```python
try:

    query = "What is Artificial Intelligence?"

    results = vectorstore.similarity_search(query, k=3)

    print("Top Retrieved Chunks:\n")

    for i, doc in enumerate(results, 1):
        print("=" * 80)
        print(f"Chunk {i}")
        print(doc.page_content[:500])
        print()

except Exception as e:
    print("❌ Retrieval Error:", e)
```
