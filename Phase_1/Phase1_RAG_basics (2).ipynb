```python
sentences = [
    "Cats are popular pets.",
    "Dogs are loyal animals.",
    "Artificial Intelligence is transforming education.",
    "Machine learning is a subset of AI.",
    "Deep learning uses neural networks.",
    "Kochi is a city in Kerala.",
    "Kerala is known for backwaters.",
    "Python is a programming language.",
    "Java is widely used in enterprise software.",
    "The Earth revolves around the Sun.",
    "Mars is called the Red Planet.",
    "Data science uses statistics and programming.",
    "Neural networks learn patterns from data.",
    "Football is a popular sport.",
    "Cricket is widely played in India.",
    "Cloud computing provides scalable resources.",
    "Databases store and manage data.",
    "RAG combines retrieval and generation.",
    "Embeddings convert text into vectors.",
    "Vector databases store embeddings."
]

print("Total Sentences:", len(sentences))
```

```python
import google.generativeai as genai
from google.colab import userdata

genai.configure(api_key=userdata.get("GEMINI_KEY"))

embeddings = []

for sentence in sentences:
    result = genai.embed_content(
        model="models/gemini-embedding-001",
        content=sentence
    )
    embeddings.append(result["embedding"])

print("Number of embeddings:", len(embeddings))
print("Dimension of first embedding:", len(embeddings[0]))
```

```python
!pip install -q scikit-learn
```

```python
from sklearn.metrics.pairwise import cosine_similarity

query = "How does AI help in education?"

# Create embedding for the query
query_embedding = genai.embed_content(
    model="models/gemini-embedding-001",
    content=query
)["embedding"]

# Compare query with all sentence embeddings
scores = cosine_similarity(
    [query_embedding],
    embeddings
)[0]

# Get Top 5 most similar sentences
top_indices = scores.argsort()[-5:][::-1]

print("Query:", query)
print("\nTop 5 Similar Sentences:\n")

for i in top_indices:
    print(f"Score: {scores[i]:.4f}")
    print(f"Sentence: {sentences[i]}")
    print()
```

# Retrieval-Augmented Generation (RAG)

Retrieval-Augmented Generation (RAG) is an AI architecture that combines information retrieval with large language models (LLMs). Instead of relying only on the model's training data, RAG retrieves relevant information from external knowledge sources and provides it as context to the LLM before generating a response.

The RAG pipeline typically consists of document loading, text extraction, chunking, embedding generation, vector database storage, retrieval of relevant chunks, and answer generation using an LLM.

Embeddings are numerical vector representations of text that capture semantic meaning. User queries and document chunks are converted into embeddings and compared using similarity measures such as cosine similarity. The most relevant chunks are retrieved and supplied to the LLM.

Advantages of RAG include improved accuracy, reduced hallucinations, access to up-to-date information, and the ability to answer questions based on custom documents.

RAG is widely used in document question answering systems, chatbots, enterprise search, knowledge management systems, and AI assistants.
