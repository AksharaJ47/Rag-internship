```python
import pickle

try:
    with open("valid_texts.pkl", "rb") as f:
        valid_texts = pickle.load(f)

    print("Loaded texts:", len(valid_texts))

except FileNotFoundError:
    print("Error: valid_texts.pkl file not found. Please upload the file.")

except pickle.UnpicklingError:
    print("Error: The file is corrupted or not a valid pickle file.")

except Exception as e:
    print("Unexpected Error:", e)
```

```python
from sentence_transformers import SentenceTransformer
import chromadb

try:
    # Load embedding model
    model = SentenceTransformer("all-MiniLM-L6-v2")

    # Create ChromaDB client
    client = chromadb.PersistentClient(path="./chroma_db")

    # Create collection
    collection = client.get_or_create_collection("phase2_docs")

    print("Model and ChromaDB loaded successfully")

except Exception as e:
    print("Initialization Error:", e)
```

```python
try:
    # Create embeddings for first 20 texts
    embeddings = model.encode(valid_texts[:20]).tolist()

    # Add to collection
    collection.add(
        documents=valid_texts[:20],
        embeddings=embeddings,
        ids=[f"doc_{i}" for i in range(20)]
    )

    print("Documents stored successfully")
    print("Total Documents:", collection.count())

except Exception as e:
    print("Storage Error:", e)
```

```python
def retrieve_documents(question):
    try:
        # Convert question into embedding
        query_embedding = model.encode(question).tolist()

        # Search similar documents
        results = collection.query(
            query_embeddings=[query_embedding],
            n_results=3
        )

        return results["documents"][0]

    except Exception as e:
        print("Retrieval Error:", e)
        return []
```

```python
question = "What is artificial intelligence?"

docs = retrieve_documents(question)

for i, doc in enumerate(docs, 1):
    print(f"\nResult {i}\n")
    print(doc[:300])
    print("\n" + "="*40)
```

```python
def rag_pipeline(question):
    try:
        print("User Question:", question)

        # Retrieve relevant documents
        docs = retrieve_documents(question)

        if len(docs) == 0:
            return "No relevant documents found."

        print("\nRetrieved Documents:\n")

        for i, doc in enumerate(docs, 1):
            print(f"Document {i}:")
            print(doc[:200])
            print("-" * 50)

        return "\nPipeline Executed Successfully"

    except Exception as e:
        return f"Pipeline Error: {e}"
```

```python
result = rag_pipeline("What is artificial intelligence?")

print(result)
```

```python
result = rag_pipeline("What is artificial intelligence?")
print(result)
```

```python
from typing import TypedDict
from langgraph.graph import StateGraph, END

class GraphState(TypedDict):
    question: str
    documents: list

def retrieve_node(state):
    docs = retrieve_documents(state["question"])
    return {"documents": docs}

workflow = StateGraph(GraphState)

workflow.add_node("retrieve", retrieve_node)

workflow.set_entry_point("retrieve")
workflow.add_edge("retrieve", END)

app = workflow.compile()
```

```python
result = app.invoke(
    {"question": "What is artificial intelligence?"}
)

print(result)
```
