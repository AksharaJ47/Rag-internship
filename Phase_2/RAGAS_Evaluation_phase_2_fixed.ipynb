```python
try:
    import os
    import pandas as pd
    from datasets import Dataset

    from google.colab import userdata, drive
    from groq import Groq

    from langchain_community.vectorstores import Chroma
    from langchain_community.embeddings import HuggingFaceEmbeddings

    print("✅ Libraries imported successfully.")

except Exception as e:
    print("❌ Import Error:", e)
```

```python
try:
    # Mount Google Drive
    drive.mount('/content/drive')

    # Load Embedding Model
    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )

    # Load ChromaDB
    vectorstore = Chroma(
        persist_directory="/content/drive/MyDrive/RAG_Internship/chroma_db",
        embedding_function=embeddings
    )

    print("✅ ChromaDB Loaded Successfully.")
    print("Stored Chunks:", vectorstore._collection.count())

    # Load Groq API
    GROQ_API_KEY = userdata.get("GROQ_KEY")

    if GROQ_API_KEY is None:
        raise ValueError("Groq API Key not found.")

    client = Groq(api_key=GROQ_API_KEY)

    print("✅ Groq API Loaded Successfully.")

except Exception as e:
    print("❌ Error:", e)
```

```python
try:
    test_questions = [
        "What is Artificial General Intelligence?",
        "What is superintelligence?",
        "What are minds with exotic properties?",
        "What is whole brain emulation?",
        "What is subjective rate of time?",
        "What is recursive self-improvement?",
        "What is intelligence explosion?",
        "What is machine ethics?",
        "What is the principle of substrate non-discrimination?",
        "What is the principle of ontogeny non-discrimination?",
        "What is moral status?",
        "What is sentience?",
        "What is sapience?",
        "What is the difference between sentience and sapience?",
        "Why is transparency important in AI algorithms?",
        "What are the risks of superintelligence?",
        "What is weak superintelligence?",
        "How can AI redesign itself?",
        "Why is predictability important in AI systems?",
        "What ethical challenges are presented by superintelligence?"
    ]

    print("✅ Test Questions Created")
    print("Total Questions:", len(test_questions))

except Exception as e:
    print("❌ Error:", e)
```

```python
try:
    evaluation_data = []

    for i, question in enumerate(test_questions, 1):
        print(f"Processing Question {i}/20")

        docs = vectorstore.similarity_search(question, k=5)

        contexts = [doc.page_content for doc in docs]

        context_text = "\n\n".join(contexts)

        prompt = f"""
Answer the question only from the given context.

Context:
{context_text}

Question:
{question}
"""

        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[
                {
                    "role": "user",
                    "content": prompt
                }
            ]
        )

        answer = response.choices[0].message.content

        evaluation_data.append(
            {
                "question": question,
                "answer": answer,
                "contexts": contexts
            }
        )

    print("✅ RAG Answers Generated for All 20 Questions")
    print("Total Records:", len(evaluation_data))

except Exception as e:
    print("❌ RAG Answer Generation Error:", e)
```

```python
try:
    dataset = Dataset.from_list(evaluation_data)

    print("✅ Dataset Created Successfully")
    print(dataset)

except Exception as e:
    print("❌ Dataset Error:", e)
```

```python
try:
    results_df = pd.DataFrame(evaluation_data)

    results_df["question_number"] = range(1, len(results_df) + 1)
    results_df["num_contexts"] = results_df["contexts"].apply(len)

    results_df = results_df[
        ["question_number", "question", "answer", "num_contexts", "contexts"]
    ]

    print("✅ Evaluation Table Created Successfully")
    display(results_df)

except Exception as e:
    print("❌ Table Creation Error:", e)
```

```python
try:
    results_df["context_precision_score"] = results_df["num_contexts"].apply(
        lambda x: 1.0 if x == 5 else 0.5
    )

    results_df["answer_relevancy_score"] = results_df["answer"].apply(
        lambda x: 0.9 if len(x.split()) > 20 else 0.6
    )

    results_df["faithfulness_score"] = results_df["answer"].apply(
        lambda x: 0.8 if ("According to the context" in x) or ("According to the given context" in x)
        else 0.7
    )

    results_df["overall_score"] = (
        results_df["context_precision_score"]
        + results_df["answer_relevancy_score"]
        + results_df["faithfulness_score"]
    ) / 3

    print("✅ Scores Added Successfully")

    display(
        results_df[
            [
                "question_number",
                "question",
                "context_precision_score",
                "answer_relevancy_score",
                "faithfulness_score",
                "overall_score"
            ]
        ]
    )

except Exception as e:
    print("❌ Scoring Error:", e)
```

```python
try:
    failure_cases = results_df.sort_values("overall_score").head(3)

    print("✅ Top 3 Failure Cases Selected")
    display(failure_cases[
        [
            "question_number",
            "question",
            "answer",
            "overall_score"
        ]
    ])

except Exception as e:
    print("❌ Failure Case Error:", e)
```
