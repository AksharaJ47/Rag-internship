```python
try:
    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )

    print("✅ Embedding Model Loaded Successfully.")

except Exception as e:
    print("❌ Embedding Model Error:", e)
```

```python
try:
    from google.colab import drive

    drive.mount('/content/drive')

    print("✅ Google Drive Mounted Successfully.")

except Exception as e:
    print("❌ Drive Mount Error:", e)
```

```python
try:
    from langchain_community.vectorstores import Chroma

    vectorstore = Chroma(
        persist_directory="/content/drive/MyDrive/RAG_Internship/chroma_db",
        embedding_function=embeddings
    )

    print("✅ ChromaDB Loaded Successfully.")
    print("Stored:", vectorstore._collection.count())

except Exception as e:
    print("❌ ChromaDB Loading Error:", e)
```

```python
try:
    question = input("Enter your question: ")

    if question.strip() == "":
        raise ValueError("Question cannot be empty.")

    print("✅ Question Received")

except Exception as e:
    print("❌ Input Error:", e)
```

```python
try:
    docs = vectorstore.similarity_search(
        question,
        k=3
    )

    context = "\n\n".join(
        [doc.page_content for doc in docs]
    )

    print("✅ Retrieved Chunks Successfully\n")
    print(context[:1000])

except Exception as e:
    print("❌ Retrieval Error:", e)
```

```python
try:
    from google.colab import userdata
    from groq import Groq
    import os

    GROQ_KEY = userdata.get("GROQ_KEY")

    if GROQ_KEY is None:
        raise ValueError("Groq API Key not found.")

    client = Groq(
        api_key=GROQ_KEY
    )

    print("✅ Groq API Loaded Successfully.")

except Exception as e:
    print("❌ Groq API Error:", e)
```

```python
try:
    prompt = f"""
Answer the question only from the given context.

Context:
{context}

Question:
{question}
"""

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ]
    )

    answer = response.choices[0].message.content

    print("\n✅ Answer Generated Successfully\n")
    print(answer)

except Exception as e:
    print("❌ Answer Generation Error:", e)
```
