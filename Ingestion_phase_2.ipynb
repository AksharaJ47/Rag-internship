```python
from langchain_community.document_loaders import PyPDFLoader

all_docs = []

for pdf in uploaded.keys():
    loader = PyPDFLoader(pdf)
    docs = loader.load()
    all_docs.extend(docs)

print("Total Pages Loaded:", len(all_docs))
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = splitter.split_documents(all_docs)

print("Total Chunks Created:", len(chunks))
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

print("Embedding model ready")
```

```python
# Convert chunks into clean text

texts = []

for chunk in chunks:
    text = str(chunk.page_content)

    if text.strip():
        texts.append(text)

print("Valid Text Chunks:", len(texts))
print("Sample:")
print(texts[0][:200])
```

```python
clean_texts = []

for t in texts[:500]:
    try:
        s = str(t)
        if len(s.strip()) > 0:
            clean_texts.append(s)
    except:
        pass

print("Clean texts:", len(clean_texts))
print("Type of first item:", type(clean_texts[0]))
print(clean_texts[0][:200])
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

test_embedding = model.encode(clean_texts[0])

print("Embedding length:", len(test_embedding))
```

```python
for i, item in enumerate(clean_texts):
    if not isinstance(item, str):
        print("Non-string at index:", i, type(item))
        break
else:
    print("All items are strings")
```

```python
for i, item in enumerate(clean_texts[:20]):
    print(i, type(item), repr(item)[:100])
```

```python
fixed_texts = [str(x).strip() for x in clean_texts]

print(type(fixed_texts[0]))
print(len(fixed_texts))
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

emb = model.encode(fixed_texts[:5])

print("Success", len(emb))
```

```python
valid_texts = []

for text in fixed_texts:
    try:
        model.encode(text)
        valid_texts.append(text)

        if len(valid_texts) == 100:
            break
    except:
        pass

print("Valid texts collected:", len(valid_texts))
```

```python
embeddings_data = model.encode(valid_texts)

print("Embeddings created:", len(embeddings_data))
```

```python
import chromadb

client = chromadb.PersistentClient(path="./chroma_db")

collection = client.get_or_create_collection(
    name="phase2_docs"
)

collection.add(
    documents=valid_texts,
    embeddings=embeddings_data.tolist(),
    ids=[f"doc_{i}" for i in range(len(valid_texts))]
)

print("ChromaDB created successfully")
print("Documents stored:", len(valid_texts))
```
