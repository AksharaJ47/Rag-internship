```python
try:
    loader = PyPDFLoader(pdf_path)
    documents = loader.load()

    print("✅ PDF Loaded Successfully")
    print("Total Pages:", len(documents))

    improved_splitter = RecursiveCharacterTextSplitter(
        chunk_size=900,
        chunk_overlap=200
    )

    improved_chunks = improved_splitter.split_documents(documents)

    print("✅ Better Chunking Completed")
    print("Improved Total Chunks:", len(improved_chunks))

except Exception as e:
    print("❌ Better Chunking Error:", e)
```

```python
try:
    improved_vectorstore = Chroma.from_documents(
        documents=improved_chunks,
        embedding=embeddings,
        persist_directory="/content/drive/MyDrive/RAG_Internship/chroma_db_phase4"
    )

    print("✅ Improved ChromaDB Created Successfully")
    print("Improved Stored Chunks:", improved_vectorstore._collection.count())

except Exception as e:
    print("❌ Improved ChromaDB Error:", e)
```

```python
try:
    chunk_texts = [doc.page_content for doc in improved_chunks]

    tokenized_chunks = [
        text.split()
        for text in chunk_texts
    ]

    bm25 = BM25Okapi(tokenized_chunks)

    print("✅ BM25 Retriever Created Successfully")
    print("Total Documents Indexed:", len(chunk_texts))

except Exception as e:
    print("❌ BM25 Error:", e)
```

```python
try:
    def hybrid_search(query, vector_k=5, bm25_k=5, final_k=5):
        # Vector search from improved ChromaDB
        vector_docs = improved_vectorstore.similarity_search(
            query,
            k=vector_k
        )

        # BM25 keyword search
        tokenized_query = query.split()
        bm25_scores = bm25.get_scores(tokenized_query)

        top_bm25_indices = sorted(
            range(len(bm25_scores)),
            key=lambda i: bm25_scores[i],
            reverse=True
        )[:bm25_k]

        bm25_docs = [
            improved_chunks[i]
            for i in top_bm25_indices
        ]

        # Combine vector + BM25 results and remove duplicates
        combined_docs = []
        seen_texts = set()

        for doc in vector_docs + bm25_docs:
            text = doc.page_content

            if text not in seen_texts:
                combined_docs.append(doc)
                seen_texts.add(text)

            if len(combined_docs) >= final_k:
                break

        return combined_docs

    print("✅ Hybrid Search Function Created Successfully")

except Exception as e:
    print("❌ Hybrid Search Function Error:", e)
```

```python
query = "What is superintelligence?"

results = hybrid_search(query)

print("Retrieved Chunks:", len(results))

for i, doc in enumerate(results, 1):
    print("\n")
    print("="*50)
    print(f"Chunk {i}")
    print("="*50)
    print(doc.page_content[:500])
```

```python
try:
    def improved_rag_answer(question):
        try:
            if not question or not isinstance(question, str):
                raise ValueError("Question must be a non-empty string.")

            docs = hybrid_search(
                question,
                vector_k=5,
                bm25_k=5,
                final_k=5
            )

            if not docs:
                raise ValueError("No relevant chunks retrieved.")

            context = "\n\n".join(
                [doc.page_content for doc in docs]
            )

            prompt = f"""
Answer the question only using the context below.

Context:
{context}

Question:
{question}

Answer:
"""

            response = client.chat.completions.create(
                model="llama-3.3-70b-versatile",
                messages=[
                    {
                        "role": "user",
                        "content": prompt
                    }
                ],
                temperature=0
            )

            answer = response.choices[0].message.content

            return {
                "question": question,
                "answer": answer,
                "retrieved_contexts": [
                    doc.page_content for doc in docs
                ]
            }

        except Exception as inner_error:
            return {
                "question": question,
                "answer": f"Error generating answer: {inner_error}",
                "retrieved_contexts": []
            }

    print("✅ Improved RAG answer function created successfully")

except Exception as e:
    print("❌ Function Creation Error:", e)
```

```python
try:
    test_question = "What is superintelligence?"

    result = improved_rag_answer(test_question)

    print("✅ Improved RAG Test Completed")
    print("\nQuestion:")
    print(result["question"])

    print("\nAnswer:")
    print(result["answer"])

    print("\nRetrieved Chunks:", len(result["retrieved_contexts"]))

except Exception as e:
    print("❌ Improved RAG Test Error:", e)
```
