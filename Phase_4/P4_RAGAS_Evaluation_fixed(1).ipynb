```python
try:
    from langchain_community.vectorstores import Chroma
    from langchain_community.embeddings import HuggingFaceEmbeddings

    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )

    improved_vectorstore = Chroma(
        persist_directory="/content/drive/MyDrive/RAG_Internship/chroma_db_phase4",
        embedding_function=embeddings
    )

    print("✅ Phase 4 ChromaDB loaded")
    print("Stored Chunks:", improved_vectorstore._collection.count())

except Exception as e:
    print("❌ Phase 4 ChromaDB Load Error:", e)
```

```python
try:
    chroma_data = improved_vectorstore._collection.get(
        include=["documents"]
    )

    chunk_texts = chroma_data["documents"]

    if not chunk_texts:
        raise ValueError("No documents found in Phase 4 ChromaDB")

    tokenized_chunks = [
        text.split()
        for text in chunk_texts
    ]

    bm25 = BM25Okapi(tokenized_chunks)

    print("✅ BM25 Retriever Recreated Successfully")
    print("Total Documents Indexed:", len(chunk_texts))

except Exception as e:
    print("❌ BM25 Creation Error:", e)
```

```python
try:
    def hybrid_search_phase4(query, vector_k=5, bm25_k=5, final_k=5):
        try:
            if not query or not isinstance(query, str):
                raise ValueError("Query must be a non-empty string")

            # Vector search
            vector_docs = improved_vectorstore.similarity_search(
                query,
                k=vector_k
            )

            # BM25 search
            tokenized_query = query.split()
            bm25_scores = bm25.get_scores(tokenized_query)

            top_bm25_indices = sorted(
                range(len(bm25_scores)),
                key=lambda i: bm25_scores[i],
                reverse=True
            )[:bm25_k]

            bm25_docs = [
                type(vector_docs[0])(
                    page_content=chunk_texts[i],
                    metadata={"source": "bm25"}
                )
                for i in top_bm25_indices
            ]

            # Combine and remove duplicates
            combined_docs = []
            seen_texts = set()

            for doc in vector_docs + bm25_docs:
                text = doc.page_content

                if text not in seen_texts:
                    combined_docs.append(doc)
                    seen_texts.add(text)

                if len(combined_docs) >= final_k:
                    break

            return combined_docs

        except Exception as inner_error:
            print("❌ Hybrid Search Inner Error:", inner_error)
            return []

    print("✅ Phase 4 Hybrid Search Function Created Successfully")

except Exception as e:
    print("❌ Hybrid Search Function Error:", e)
```

```python
try:
    def improved_rag_answer_phase4(question):
        try:
            docs = hybrid_search_phase4(
                question,
                vector_k=5,
                bm25_k=5,
                final_k=5
            )

            if not docs:
                raise ValueError("No chunks retrieved")

            context = "\n\n".join(
                [doc.page_content for doc in docs]
            )

            prompt = f"""
Answer the question only from the given context.

Context:
{context}

Question:
{question}

Answer:
"""

            response = client.chat.completions.create(
                model="llama-3.3-70b-versatile",
                messages=[
                    {
                        "role": "user",
                        "content": prompt
                    }
                ],
                temperature=0
            )

            answer = response.choices[0].message.content

            return {
                "question": question,
                "answer": answer,
                "retrieved_contexts": [
                    doc.page_content for doc in docs
                ]
            }

        except Exception as inner_error:
            return {
                "question": question,
                "answer": f"Error: {inner_error}",
                "retrieved_contexts": []
            }

    print("✅ Phase 4 improved answer function created")

except Exception as e:
    print("❌ Function Creation Error:", e)
```
