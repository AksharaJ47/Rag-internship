```python
try:
    phase2_df = pd.read_csv("ragas_results_ethics_ai.csv")
    phase4_df = pd.read_csv("phase4_ragas_results (3).csv")

    phase2_score = phase2_df["context_precision"].dropna().mean()
    phase4_score = phase4_df["context_precision"].dropna().mean()

    comparison_df = pd.DataFrame({
        "Metric": ["Context Precision"],
        "Phase 2 Score": [phase2_score],
        "Phase 4 Score": [phase4_score],
        "Improvement": [phase4_score - phase2_score],
        "Note": [
            "Phase 4 evaluated using valid Context Precision scores. Other values returned NaN due to Groq rate limit."
        ]
    })

    print("✅ Comparison created")
    display(comparison_df)

except Exception as e:
    print("❌ Comparison Error:", e)
```

```python
try:
    comparison_df.to_csv(
        "phase2_vs_phase4_comparison.csv",
        index=False
    )

    print("✅ phase2_vs_phase4_comparison.csv saved successfully")

except Exception as e:
    print("❌ CSV Save Error:", e)
```
